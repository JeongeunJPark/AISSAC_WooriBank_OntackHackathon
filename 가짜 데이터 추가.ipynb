```python
# 시그모이드 함수
def sigmoid(x):
    return 1/(1+np.exp(-x).astype(np.float64))
```

```python
def tanh(x):  # 하이퍼볼릭 탄젠트 함수
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
def relu(x):  # relu 함수
    return max(0,x)
```
